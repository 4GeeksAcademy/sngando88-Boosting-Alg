```python
# Your code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/decision-tree-project-tutorial/main/diabetes.csv')
```

```python
main_df
```

```python
X = main_df.drop(columns='Outcome')
y = main_df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
X
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train
```

```python
gra_boost_model = GradientBoostingClassifier(learning_rate=0.0001, n_estimators=100)
```

```python
gra_boost_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
def scoring_make(model, X_train, X_test, y_train, y_test):
    y_train_preds = model.predict(X_train)
    y_test_preds = model.predict(X_test)

    print(f'accuracy of training dataset is {accuracy_score(y_true=y_train, y_pred=y_train_preds)}')
    print(f'accuracy of testing dataset is {accuracy_score(y_true=y_test, y_pred=y_test_preds)}')

    print('confusion matrix of training dataset is\n')
    print(confusion_matrix(y_true=y_train, y_pred=y_train_preds))
    print('confusion matrix of testing dataset is\n')
    print(confusion_matrix(y_true=y_test, y_pred=y_test_preds))

    print(f'classification report of training dataset is\n')
    print(classification_report(y_true=y_train, y_pred=y_train_preds))
    print(f'classification report of testing dataset is\n')
    print(classification_report(y_true=y_test, y_pred=y_test_preds))
```

```python
scoring_make(gra_boost_model, X_train, X_test, y_train, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyperparameters = {
    "learning_rate" : [0.1,0.2,0.3,0.4],
    "n_estimators" : [20, 50, 100, 150],
    "max_depth" : [2, 5, 8, 10],
    
}
```

```python
grid_model = GridSearchCV(GradientBoostingClassifier(random_state=25), param_grid= hyperparameters, scoring="accuracy", cv = 5)

grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
best_gra_model = GradientBoostingClassifier(learning_rate = 0.1, max_depth = 2, n_estimators= 50)
```

```python
best_gra_model.fit(X_train, y_train)
```

```python
scoring_make(best_gra_model, X_train, X_test, y_train, y_test)
```
